```python
sns.countplot(x='Ethnicity',data=df)
plt.title('Ethnicity',color='blue')
plt.show()
```

```python
sns.histplot(x='Age',data=df,color='green')
plt.title('Age')
plt.show()
```

```python
num_df=df.select_dtypes(include=['float64','int64'])
plt.figure(figsize=(10,10))
sns.heatmap(num_df.corr(),annot=True)
plt.show()
```

```python
sns.scatterplot(x='Age',y='Gender',data=df,markers='o',color='red')
plt.xlabel('Age')
plt.ylabel('Gender')
plt.title('Age vs Gender')
plt.show()
```

```python
sns.lineplot(x='Sports',y='Music',data=df,color='green')
plt.xlabel('Sports')
plt.ylabel('Music')
plt.title('Sports vs Music')
plt.show()
```
